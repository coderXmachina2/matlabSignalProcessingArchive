```python
import datetime
import requests
import urllib

                                                   #Initialize Date
theTime = datetime.datetime(2013, 4, 4, 1, 44, 34) #year month day hour, minute, second 

#
#set naming index
index0 = 0
index1 = 7

print("Starting scrape operation: " + str(theTime) + " with index " + str(index0) + str(index1) + "\n")

#start iteration
for l in range(0,94): #download one day
    
    for p in range(0,150): #cycle through these seconds for a possible new image download
        
        #form download string
        theTime =  theTime + datetime.timedelta(0,1) #start increment at 1
        repString = str(theTime.year)+"/"+str(format(theTime.month, '02'))+"/"+str(format(theTime.day, '02'))+"/"+str(theTime.year)+str(format(theTime.month, '02'))+str(format(theTime.day, '02'))
        tReq = str(format(theTime.hour, '02')) + str(format(theTime.minute, '02')) + str(format(theTime.second, '02'))
        
        #this is the download String
        downloadString = "https://sdo.gsfc.nasa.gov/assets/img/browse/"+repString+"_" + tReq + "_1024_0171.jpg"
        
        request = requests.get(downloadString)
        if request.status_code == 200: #valid website
            print('Image at ' + downloadString +  ' exists. Downloading now.')
            folderString = "praiseTheSun/20130404/" + str(index0) + str(index1) + ".jpg" #save directory
            myImage = urllib.request.urlretrieve(downloadString, folderString) #download successful
            print("Downloaded image " + str(index0) + str(index1) + "\n")
            index1 = index1 + 1
            theTime  = theTime + datetime.timedelta(0,790) #minimum jump 13 minutes 30 seconds
            
            break;

        else:
            print(str(p) + " Image at " + tReq + " does not exist. Increment")
```
